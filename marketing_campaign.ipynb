Data Cleaning and Analysis Using KLIB

```python
#importing data and libraries

import klib
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Chris\Documents\Datasets\marketing_campaign.csv")
df.head()
```

Cleaning and Analysing Data



```python
#ploting missing values

klib.missingval_plot(df)
```

```python
df.describe()
```

```python
df.info()
```

Finding nulls in the data


```python
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
#removing empty and duplicate rows

df_cleaned = klib.data_cleaning(df)
```

Displaying correlation data

```python
klib.corr_plot(df_cleaned, annot=False)
```

```python
#Displaying only positive correlation

klib.corr_plot(df_cleaned, split ='pos', annot=False)
```

```python
#Displaying only negative correlation

klib.corr_plot(df_cleaned, split ='neg', annot=False)
```

```python
#Displaying correlation with a target variable of interest

klib.corr_plot(df_cleaned, target= 'amount')
```

Categorical Plot

```python

klib.cat_plot(df,figsize= (50,15))
```

```python
plot = klib.corr_plot(df_cleaned, annot= False, figsize= (12,10))
```

```python
#saving plot

plot.figure.savefig('figure.pdf')
```
